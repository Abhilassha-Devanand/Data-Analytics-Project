```python
def fix_match_winner(row):
    if row['second_ings_score'] > row['first_ings_score']:
        return row['team_2']
    else:
        return row['team_1']

df.loc[df['match_winner'] == 'm', 'match_winner'] = (
    df[df['match_winner'] == 'm'].apply(fix_match_winner, axis=1)
)
```

```python
matches_played = (
    pd.concat([df['team_1'], df['team_2']])
    .value_counts()
    .reset_index()
)

matches_played.columns = ['team', 'matches_played']
```

```python
matches_won = (
    df['match_winner']
    .value_counts()
    .reset_index()
)

matches_won.columns = ['team', 'matches_won']
```

```python
team_performance = matches_played.merge(
    matches_won, on='team', how='left'
)

team_performance['matches_won'] = team_performance['matches_won'].fillna(0)

# for win percentage
team_performance['win_percentage'] = (
    team_performance['matches_won'] / team_performance['matches_played']
) * 100

team_performance.sort_values(
    'win_percentage', ascending=False
)
```
