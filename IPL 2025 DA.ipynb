```python
def fix_match_winner(row):
    if row['second_ings_score'] > row['first_ings_score']:
        return row['team_2']
    else:
        return row['team_1']

df.loc[df['match_winner'] == 'm', 'match_winner'] = (
    df[df['match_winner'] == 'm'].apply(fix_match_winner, axis=1)
)
```

```python
matches_played = (
    pd.concat([df['team_1'], df['team_2']])
    .value_counts()
    .reset_index()
)

matches_played.columns = ['team', 'matches_played']
```

```python
matches_won = (
    df['match_winner']
    .value_counts()
    .reset_index()
)

matches_won.columns = ['team', 'matches_won']
```

```python
team_performance = matches_played.merge(
    matches_won, on='team', how='left'
)

team_performance['matches_won'] = team_performance['matches_won'].fillna(0)

# for win percentage
team_performance['win_percentage'] = (
    team_performance['matches_won'] / team_performance['matches_played']
) * 100

team_performance.sort_values(
    'win_percentage', ascending=False
)
```

```python
toss_impact = (
    df['toss_and_match_win']
    .value_counts(normalize=True)
    * 100
).reset_index()

toss_impact.columns = ['toss_and_match_win', 'percentage']
toss_impact
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_theme(style="whitegrid")

# 1. Team Wins Bar Chart
plt.figure(figsize=(12, 7))
wins = df['match_winner'].value_counts().sort_values(ascending=False)
sns.barplot(x=wins.values, y=wins.index, hue=wins.index, palette='viridis', legend=False)
plt.title('Total Wins by Team - IPL 2025', fontsize=16)
plt.xlabel('Number of Wins', fontsize=12)
plt.ylabel('Team', fontsize=12)
plt.tight_layout()
plt.savefig('team_wins_analysis.png')
plt.s()
```

```python
# 2. Average First Innings Score by Venue
plt.figure(figsize=(12, 7))
venue_scores = df.groupby('venue')['first_ings_score'].mean().sort_values(ascending=False)
# Removing potential outliers if score < 50 (like the '7' runs match) for a better average representation
clean_venue_scores = df[df['first_ings_score'] > 50].groupby('venue')['first_ings_score'].mean().sort_values(ascending=False)
sns.barplot(x=clean_venue_scores.values, y=clean_venue_scores.index, hue=clean_venue_scores.index, palette='magma', legend=False)
plt.title('Average First Innings Score per Venue', fontsize=16)
plt.xlabel('Average Score', fontsize=12)
plt.ylabel('Venue', fontsize=12)
plt.tight_layout()
plt.savefig('venue_scoring_trends.png')
plt.show()
```

```python
# 3. Toss Decision vs Match Outcome
toss_match_win = df[df['toss-winner'] == df['match_winner']]
toss_match_loss = df[df['toss-winner'] != df['match_winner']]

labels = ['Toss Winner Won Match', 'Toss Winner Lost Match']
counts = [len(toss_match_win), len(toss_match_loss)]

plt.figure(figsize=(8, 8))
plt.pie(counts, labels=labels, autopct='%1.1f%%', colors=['#4CAF50', '#FF5252'], startangle=140, explode=(0.05, 0))
plt.title('Impact of Winning the Toss on Match Result', fontsize=16)
plt.tight_layout()
plt.savefig('toss_impact_analysis.png')
plt.show()
```

```python
# 4. Top 10 Individual High Scores
plt.figure(figsize=(12, 7))
top_scores = df.nlargest(10, 'highscore')[['top_scorer', 'highscore']]
sns.barplot(data=top_scores, x='highscore', y='top_scorer', hue='top_scorer', palette='rocket', legend=False)
plt.title('Top 10 Individual High Scores of the Season', fontsize=16)
plt.xlabel('Runs Scored', fontsize=12)
plt.ylabel('Player', fontsize=12)
plt.tight_layout()
plt.savefig('top_individual_performances.png')
plt.show()
```

```python
# 5. Winning Margin Analysis (Runs)
plt.figure(figsize=(10, 6))
run_margins = df[df['won_by'] == 'Runs']['margin'].dropna()
sns.histplot(run_margins, bins=15, kde=True, color='blue')
plt.title('Distribution of Win Margins (by Runs)', fontsize=16)
plt.xlabel('Margin (Runs)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.savefig('win_margin_runs.png')
plt.show()
```

```python
# 6. Player of the Match Leaders
plt.figure(figsize=(12, 7))
potm = df['player_of_the_match'].value_counts().head(10)
sns.barplot(x=potm.values, y=potm.index, hue=potm.index, palette='coolwarm', legend=False)
plt.title('Top 10 Player of the Match Award Winners', fontsize=16)
plt.xlabel('Number of Awards', fontsize=12)
plt.ylabel('Player', fontsize=12)
plt.tight_layout()
plt.savefig('potm_leaders.png')
plt.show()
```
